### Descrição dos Atributos

- **case**: Um número que denota um páis específico.
- **cc3**: Código de 3 letras que denota um país específico.
- **country**: Nome do país.
- **year**: Ano da observação.
- **systemic_crisis**: 0 significa que nenhuma crise sistemica ocorreu naquele ano, 1 significa que ocorreu uma crise sistemica naquele ano.
- **exch_usd**: Taxa de câmbio do país em relação ao dólar.
- **domestic_debt_in_default**: Indica se houve um incumprimento da divida interna soberana naquele ano (0=não; 1=sim).
- **sovereign_external_debt_default**: Indica se houve um incumprimento da divida externa soberana naquele ano (0=não; 1=sim).
- **gdp_weighted_default**: Total da dívida inadimplente em relação ao PIB.
- **inflation_annual_cpi**: Taxa anual de inflação do IPC.
- **independence**: Indica se um país é independente ou não.
- **currency_crises**: Indica se ocorreu ou não uma crise cambial em determinado ano. (0=não; 1=sim).
- **inflation_crises**: Indica se ocorreu ou não uma crise de inflação em determinado ano. (0=não; 1=sim).
- **banking_crisis**: Indica se ocorreu ou não uma crise bancária em determinado ano. (0=não; 1=sim).

### Objetivo
> Prever a ocorrência de crises bancárias (`banking_crisis`) de um país em determinado ano com demais informações.

### Baseline
> Superar acurácia de 85% de um modelo pre-existente.

## Visualizando os dados

### Taxa de Câmbio do Páis em Relação ao Dólar

```python
plt.figure(figsize=(20, 20))
count = 1

for country in countries:
    plt.subplot(5, 3, count)
    count += 1
    # col = '#' + ''.join([random.choice()])
    sns.lineplot(data[data.country == country]['year'], data[data.country == country]['exch_usd'], label=country)
    plt.plot([np.min(data[np.logical_and(data.country==country,data.independence==1)]['year']),
              np.min(data[np.logical_and(data.country==country,data.independence==1)]['year'])],
             [0,
              np.max(data[data.country==country]['exch_usd'])],
             color='black',
             linestyle='dotted',
             alpha=0.8)
    plt.text(np.min(data[np.logical_and(data.country == country, data.independence == 1)]['year']),
             np.max(data[data.country == country]['exch_usd']) / 2,
             'Independência',
             rotation=-90)
    plt.title(country)

plt.tight_layout()
plt.show()
```
